```python
from simplesim import simulator
```

Basic Autoregressive Decay in Discrete Time
--

```python
init     = "x[0]=5"
dynamics = "x[t]=0.9*x[t-1]"

s=simulator(init, dynamics)
print(s)
s.simulate(stop=100, debug=False)
s.plot("x")
```

Basic Autoregressive Decay in Discrete Time Plus Random Noise
--

```python
init     = "x[0]=5, y[0]=0"
dynamics = "y[t]=n_eps/10, x[t] = 0.95*x[t-1] + y[t]"

s=simulator(init, dynamics)
s.simulate(stop=100, debug=False)
s.plot()
```

A Basic Model for a Walking Suggestion MRT
--
Steps (S) depend on motivation (M). Motivation (M) is influenced by past motivation and walking suggestions (A), which occur randomly.

```python
init     = "S[0] = 2000, M[0]=0.1, A[0]=0"
dynamics = "A[t] = (u_eps>0.5) ; \
            M[t] = 0.8*M[t-1] + 0.2*A[t]; \
            S[t] = 1000 + 8000*M[t] + 1000*u_eps"

s=simulator(init, dynamics)
s.simulate(stop=14*5, debug=False)
s.plot("S,A", ylim=(-400,10000))
s.plot("M,A", ylim=(-0.1,1.1))
```

Direct Functions of Time
--

```python
init     = "x[0]=0, y[0]=0"
dynamics = "x[t] = sin(t/100), y[t]=exp(x[t])-1"

s=simulator(init, dynamics)
s.simulate(stop=1000, debug=False)
s.plot()
```
